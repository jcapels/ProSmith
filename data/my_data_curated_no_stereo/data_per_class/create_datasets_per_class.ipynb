```python
def create_datasets_per_class(splits_path, dataset_path, output_path):
    # import pickle file

    import pickle
    with open(splits_path, 'rb') as file:
        data = pickle.load(file)

    import pandas as pd

    dataset = pd.read_csv(dataset_path)

    for compound_type in data:
        # get the compounds of the current type
        compounds = data[compound_type]

        # create a new dataframe with only the compounds of the current type
        df = dataset[dataset['molecule ID'].isin(compounds)]
        print("I'm gonna write the file for the compound type: ", f"{output_path}/{compound_type}_test.csv")
        # save the dataframe to a csv file
        df.to_csv(f"{output_path}/{compound_type}_test.csv", index=False)
```

```python
create_datasets_per_class('../../../pathway_to_compounds_split_02.pkl', "../train_val_compounds_02/test.csv", "../train_val_compounds_02")
create_datasets_per_class('../../../pathway_to_compounds_split_04.pkl', "../train_val_compounds_04/test.csv", "../train_val_compounds_04")
create_datasets_per_class('../../../pathway_to_compounds_split_06.pkl', "../train_val_compounds_06/test.csv", "../train_val_compounds_06")
create_datasets_per_class('../../../pathway_to_compounds_split_08.pkl', "../train_val_compounds_08/test.csv", "../train_val_compounds_08")
```

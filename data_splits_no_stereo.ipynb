```python
import pickle

def read_pickle(file_path):
    # Open the pickle file and load the data
    with open(file_path, 'rb') as file:
        data = pickle.load(file)
    return data
```

```python
import pandas as pd

dataset = pd.read_csv("curated_dataset_no_stereochemistry_duplicates.csv")
dataset.shape
```

```python
import os

dataset = dataset.rename(columns={"Enzyme ID": "Uniprot ID", "Substrate ID": "molecule ID", "Binding":"output", "Sequence": "Protein sequence"})

def create_datasets(file_path, folder_path, proteins_split=True):

    # Replace 'your_file.pkl' with the path to your pickle file
    data = read_pickle(file_path)

    os.makedirs(folder_path, exist_ok=True)

    train_ids, val_ids, test_ids = data[0]
    if proteins_split:
        train_dataset = dataset[dataset["Uniprot ID"].isin(train_ids)]
        val_dataset = dataset[dataset["Uniprot ID"].isin(val_ids)]
        test_dataset = dataset[dataset["Uniprot ID"].isin(test_ids)]
    else:
        train_dataset = dataset[dataset["molecule ID"].isin(train_ids)]
        val_dataset = dataset[dataset["molecule ID"].isin(val_ids)]
        test_dataset = dataset[dataset["molecule ID"].isin(test_ids)]

    train_dataset.to_csv(f"{folder_path}train.csv")
    val_dataset.to_csv(f"{folder_path}val.csv")
    test_dataset.to_csv(f"{folder_path}test.csv")

file_path = 'splits_0_2_proteins_train_val_test.pkl'
create_datasets(file_path, "data/my_data_curated_no_stereo/train_val_proteins_02/")
file_path = 'splits_0_4_proteins_train_val_test.pkl'
create_datasets(file_path, "data/my_data_curated/train_val_proteins_04/")
file_path = 'splits_0_6_proteins_train_val_test.pkl'
create_datasets(file_path, "data/my_data_curated/train_val_proteins_06/")
file_path = 'splits_0_8_proteins_train_val_test.pkl'
create_datasets(file_path, "data/my_data_curated/train_val_proteins_08/")

file_path = 'splits_compounds_02.pkl'
create_datasets(file_path, "data/my_data_curated/train_val_compounds_02/", proteins_split=False)
file_path = 'splits_compounds_04_corrected.pkl'
create_datasets(file_path, "data/my_data_curated/train_val_compounds_04/", proteins_split=False)
file_path = 'splits_compounds_06.pkl'
create_datasets(file_path, "data/my_data_curated/train_val_compounds_06/", proteins_split=False)
file_path = 'splits_compounds_08.pkl'
create_datasets(file_path, "data/my_data_curated/train_val_compounds_08/", proteins_split=False)
```
